# 蔡英文就職演說詞頻

```python
from collections import Counter

import jieba
```

```python
jieba.set_dictionary('./dict.txt.big')
```

```python
with open('./speech.txt', encoding='utf-8') as f:
    speech = f.read()
```

```python
for word in jieba.cut(speech):
    print(word)
```

```python
words = [word for word in jieba.cut(speech)]
```

```python
words
```

```python
word_cnt = Counter()
```

```python
for word in words:
    if word in word_cnt:
        word_cnt[word] += 1
    else:
        word_cnt[word] = 1
```

```python
word_cnt.most_common()
```

```python
def form2dict(form_data):
    """
    method:search
    searchMethod:true
    searchTarget:ATM
    orgName:
    orgId:
    hid_1:1
    tenderName:
    tenderId:
    tenderStatus:4,5,21,29
    tenderWay:
    awardAnnounceStartDate:106/03/16
    awardAnnounceEndDate:106/03/16
    proctrgCate:
    tenderRange:
    minBudget:
    maxBudget:
    item:
    hid_2:1
    gottenVendorName:
    gottenVendorId:
    hid_3:1
    submitVendorName:
    submitVendorId:
    location:
    priorityCate:
    isReConstruct:
    btnQuery:查詢
    """
    form_dict = {}
    rows = form_data.split('\n')
    for row in rows:
        print("[DEBUG] row is %s"%(row))
        if row:
            kv = row.strip().split(":")
            form_dict[kv[0]] = kv[1]
    return form_dict
```

## 爬蟲換頁

```python
import requests as r
from bs4 import BeautifulSoup

HOST = 'https://www.ptt.cc'

res = r.get("https://www.ptt.cc/bbs/Gossiping/index.html", headers={'cookie': "over18=1"})
```

```python
soup = BeautifulSoup(res.text, 'lxml')
```

```python
soup.select('#action-bar-container > div > div.btn-group.btn-group-paging > a')[1]['href']
```

```python
soup.select('#action-bar-container > div > div.btn-group.btn-group-paging > a')[1]['href'].split("index")[1]
```

```python
page_str = soup.select('#action-bar-container > div > div.btn-group.btn-group-paging > a')[1]['href'].split("index")[1].split(".html")[0]
```

```python
page_num = int(page_str)
```

```python
page_num
```

```python
# 產生page_num數列
for page in range(page_num, page_num - 5, -1):
    print(page)
```

```python
# 產生page_num數列
for page in range(page_num, page_num - 5, -1):
    print(page)
    res = r.get("https://www.ptt.cc/bbs/Gossiping/index{}.html".format(page), headers={'cookie': "over18=1"})
    print(res)
```

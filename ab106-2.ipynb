# 蔡英文就職演說詞頻

```python
from collections import Counter

import jieba
```

```python
jieba.set_dictionary('./dict.txt.big')
```

```python
with open('./speech.txt', encoding='utf-8') as f:
    speech = f.read()
```

```python
for word in jieba.cut(speech):
    print(word)
```

```python
words = [word for word in jieba.cut(speech)]
```

```python
words
```

```python
word_cnt = Counter()
```

```python
for word in words:
    if word in word_cnt:
        word_cnt[word] += 1
    else:
        word_cnt[word] = 1
```

```python
word_cnt.most_common()
```

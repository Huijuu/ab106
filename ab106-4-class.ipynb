# SQLAlchemy

```python
!pip install sqlalchemy
```

```python
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy import Column, String, DateTime, Text
from sqlalchemy import create_engine
```

```python
Base = declarative_base()
```

```python
dir(Base)
```

```python
class Article(Base):
    __tablename__ = 'article'
    
    url = Column(String(255), primary_key=True)
    title = Column(String(255))
    content = Column(Text)
    author = Column(String(255))
    dt = Column(String(255))
    board = Column(String(255))
    ip = Column(String(255))
```

```python
engine = create_engine("mysql+pymysql://root:ilovepython@localhost/ab106?charset=utf8mb4")
```

```python
engine
```

```python
Base.metadata.create_all(engine)
```

```python
from sqlalchemy.orm import sessionmaker
```

```python
session = sessionmaker(bind=engine)
```

```python
s = session()
```

```python
s
```

```python
ptt_articles[0]
```

```python
s.add(Article(**ptt_articles[0])) # Article(author = 'ggcip0514 (汪汪)', board = 'Gossiping', )
```

```python
s.commit()
```

```python
s.rollback()
```

```python
s.bulk_insert_mappings(Article, ptt_articles)
```

```python
s.commit()
```

```python
cur = s.query(Article)
```

```python
for row in cur:
    print(row.url, row.dt, row.title, row.ip)
```

```python
cur = s.query(Article).filter(Article.dt > datetime.datetime.strptime('2017-03-23 11:08:31', '%Y-%m-%d %H:%M:%S'))
```

```python
for row in cur:
    print(row.title, row.url, row.dt)
```
